```python
import os
import json
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sys
sys.path.append(os.path.join(os.getcwd(), "..", "scripts"))
from utils import basic_Q_features, ST_features
```

```python
result_dir = os.path.join("..", "models", "2025-12-29_sfcn++")
config_json_path = os.path.join(result_dir, "config.json")
trainval_hist_path = os.path.join(result_dir, "train-val_history.csv")
trainval_preds_path_template = os.path.join(result_dir, "train-val_preds_fold-{}.csv")
test_preds_path = os.path.join(result_dir, "test_predictions.csv")
subj_data_path = os.path.join("..", "data", "meta", "subj_Q+ST.csv")
```

```python
with open(config_json_path, 'r', errors='ignore') as f:
    config_json = json.load(f)
```

```python
class Config:
    def __init__(self, data: dict, args=None):
        if data:
            self.__dict__.update({
                k.lower(): v for k, v in data.items()
            })
```

```python
config = Config(config_json)
```

```python
trainval_hist_df = pd.read_csv(trainval_hist_path)
best_fold = trainval_hist_df.sort_values(by=[f"val_{config.metric}"], ascending=True).iloc[0]["fold"]
```

```python
trainval_preds_path = trainval_preds_path_template.format(int(best_fold))
trainval_preds = pd.read_csv(trainval_preds_path)
test_preds = pd.read_csv(test_preds_path)
```

```python
test_preds.insert(0, "Set", "Test")
all_preds = pd.concat([test_preds, trainval_preds])
all_preds.set_index("SID", inplace=True)
all_preds["PAD"] = all_preds["Pred_Age"] - all_preds["True_Age"]
```

```python
subj_data = pd.read_csv(subj_data_path)
subj_data.set_index("BASIC_INFO_ID", inplace=True)
```

```python
dat = pd.merge(all_preds, subj_data, left_index=True, right_index=True)
```

```python
# print(len(all_preds))
# print(len(subj_data))
# print(len(dat))
```

```python
basic_q_features = basic_Q_features()
st_features = ST_features()
```

```python
cormat = pd.DataFrame(index=st_features, columns=["Train", "Val", "Test"], dtype=float)
p_mat = cormat.copy(deep=True)

for f in st_features:
    for s in ["Train", "Val", "Test"]:
        targ_df = dat.query(f"Set == '{s}'").loc[:, ["PAD", f]].dropna()
        cormat.loc[f, s], p_mat.loc[f, s] = stats.pearsonr(targ_df["PAD"], targ_df[f])

p_sig = p_mat.map(lambda x: "*" * sum( x <= t for t in [0.05, 0.01, 0.001] ))
annot_mat = cormat.map(lambda x: f"{x:.2f}") + p_sig
```

```python
dat.query("Set == 'Train'")["PAD"].mean()
```

```python
sns.set_theme(style='white', font_scale=1.1)
plt.figure(figsize=(10, 8), dpi=200)
g = sns.heatmap(
    cormat, mask=None, square=False, 
    vmin=-1, vmax=1, linewidth=.5, cmap="RdBu_r", cbar=False, 
    annot=pd.DataFrame(annot_mat), fmt = "", annot_kws={"size": 20}
)
g.set(xlabel="", ylabel="")
g.tick_params(axis="both", which="major", labelsize=20)
plt.tight_layout()
```
